```python
def tran_date(df):
    for i in range(len(df['date'])):
        new_date = df['date'][i].split("T")
        new_date = new_date[0][: 4] + '-' + new_date[0][4: 6] + '-' + new_date[0][6:]
        df['date'][i] = new_date
        print(new_date)
    def parser(x):
        return datetime.datetime.strptime(x, '%Y-%m-%d')
    df['date'] = df['date'].map(lambda x: parser(x))

    return df 
```

```python
df = tran_date(df)
```

```python
def create_colum(df):

    df['weekday'] = df['date'].apply(lambda x: x.date().weekday())

    year = df['date'].values
    lst_month = []
    lst_year = []
    for i in year:
        a = str(i)
        month = a.split('-')[1]
        year = a.split('-')[0]
        lst_month.append(int(month))
        lst_year.append(int(year))
    df['month'] = lst_month
    df['year'] = lst_year

    return df

df = create_colum(df)
df.head()
```

```python
def find_outlier(df):
    features_75 = {}
    features_25 = {}

    for i in df.columns:
        q25, q50 , q75 = np.percentile(df[i], [25, 50, 75])
        media = q75 - q25
        max_ = q75 + 1.5* media
        min_ = q25 - 1.5* media

        count_75 = np.where(df[i] > max_)
        count_25 = np.where(df[i] < min_)

        value_75 = (len(count_75[0]),count_75[0])
        value_25 = (len(count_25[0]),count_25[0])

        features_75[i] = value_75
        features_25[i] = value_25   
    
    for i in features_75:
        if features_75[i][0] != 0:
            for j in features_75[i][1]:
                df[i][j] = None
    for i in features_25:
        if features_25[i][0] != 0:
            for j in features_25[i][1]:
                df[i][j] = None
```
